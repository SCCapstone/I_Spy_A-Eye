This generates JavaScript code that can validate wether a zipcode belongs to a state name or state abbreviation. The result still needs minor manual formatting and a simple automaatic code formatter like Prettier. 
Data Source:
http://www.structnet.com/instructions/zip_min_max_by_state.html

```python
import pandas as pd

pd.set_option('display.max_rows', 200)

df = pd.read_csv('state_zipcode_ranges.csv')
# So zip codes can be printed in concatenation
df = df.applymap(str)
df.head(100)
```

```python
for index, row in df.iterrows():
    print('else if ((state === "' + row['State Name'] + '" || state === "' + row['ST'] + '") && zipCode >= ' + row['Zip Min'] + ' && zipCode <= ' + row['Zip Max'] + ') {return true;}')
```
